```python
import pandas as pd # Data manipulation
import numpy as np #
import seaborn as sns # Data exploration and visualization
import matplotlib.pyplot as plt # Data visualization and graphical charting
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv("data_set\Fake.csv")
data_true = pd.read_csv("data_set\True.csv")
```

```python
data_fake.head()
```

```python
data_true.head() # Tail for last 5
```

```python
data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing = data_fake.tail(10)
data_fake = data_fake.iloc[:-10]
    
data_true_manual_testing = data_true.tail(10)
data_true = data_true.iloc[:-10]
```

```python
data_fake_manual_testing['class'] = 0
data_true_manual_testing['class'] = 1
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(['title', 'subject', 'date'], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1) # Randomizing, shuffling (frac = 1 means give me all of it, just shuffled)
```

```python
data.reset_index(inplace = True) # Resets the index, inplace = True means to do it directly to data not make a copy
data.drop(['index'], axis = 1, inplace = True) # deletes old index column
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text) # Removes things in square brackets; Like [Source: Wikipedia] → gone.
    text = re.sub("\\W", " ", text) # Removes non-word characters (symbols like @, #, etc.) and replaces them with spaces.
    text = re.sub("https?://\S+|www\.\S+", "", text) # Removes URLs
    text = re.sub("<.*?>+", "", text) # Removes HTML tags; Like <div>hello</div> → just "hello"; Sometimes news articles come with HTML code. This clears it out.
    text = re.sub('[%s]' % re.escape(string.punctuation), "", text) # Removes punctuation; Goodbye commas, periods, exclamation marks...
    text = re.sub('\n', "", text) # Removes newlines (the Enter key / line breaks)
    text = re.sub('\w*\d\w*', "", text) # Removes words that have numbers in them Like: covid19, election2024, user123 → all get removed. We remove it for less noisiness and keep vocabulary smaller; for big datasets it is okay to leave
    
    return text
```

```python
data['text'] = data['text'].apply(wordopt)
```

```python
x = data['text']
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

'''
You're using TF-IDF (Term Frequency – Inverse Document Frequency) to convert text into vectors (a list of numbers) that represent the importance of each word in the text.

Look at all the unique words in your training data
Assign a score to each word based on how often it shows up (TF) and how unique it is (IDF)
'''

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train) # Use on training data
xv_test = vectorization.transform(x_test) # Use on test data so it will not learn on this data
```

```python
from sklearn.linear_model import LogisticRegression # Great for binary classification problems

LR = LogisticRegression()
LR.fit(xv_train, y_train)
```

```python
pred_lr = LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

'''
Does the text contain the word "shocking"?
    ↳ Yes → Maybe fake
    ↳ No → Does it contain a date?
        ↳ Yes → Maybe real
        ↳ No → Maybe fake

If x_train has articles like:

“Click here! Shocking news!” → Fake

“Government passes a new law” → Real

The tree might learn:

If a word like “click” appears → predict fake.

If the article talks about official sources → predict real.        

'''

DT = DecisionTreeClassifier()
DT.fit(xv_train, y_train)
```

```python
pred_dt = DT.predict(xv_test)
```

```python
DT.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier # Builds many small decision trees, each improving on the last one's mistakes; fear of overfitting
GB = GradientBoostingClassifier(random_state = 0)
GB.fit(xv_train, y_train)
```

```python
predict_gb = GB.predict(xv_test)
```

```python
GB.score(xv_test, y_test)
```

```python
print(classification_report(y_test, predict_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier # A collection of decision trees voting together

RF = RandomForestClassifier(random_state = 0)
RF.fit(xv_train, y_train)
```

```python
pred_rf = RF.predict(xv_test)
```

```python
RF.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"
    
def manual_testing(news):
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GB = GB.predict(new_xv_test)
    pred_RF = RF.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGB Prediction: {} \nRF Prediction: {}".format(output_lable(pred_LR[0]), output_lable(pred_DT[0]), output_lable(pred_GB[0]), output_lable(pred_RF[0])))
```
